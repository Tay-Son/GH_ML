```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pycaret.classification import *
```

```python
dir_dataset = "C:/Users/0stix/Datasets/"

name_dataset = 'TPS_sep_2021'

df_train = pd.read_csv(dir_dataset+name_dataset+'/train.csv')
df_test = pd.read_csv(dir_dataset+name_dataset+'/test.csv')
df_sub = pd.read_csv(dir_dataset+name_dataset+'/sample_solution.csv')
len_train = len(df_train)
target='claim'

df_all = pd.concat([df_train,df_test],axis=0)
df_all.head()
```

```python
df_all_v1 = df_all.copy()

df_train_v1 = df_all_v1[:len_train]
df_test_v1 = df_all_v1[len_train:]
df_test_v1.drop(target,axis=1,inplace=True)

df_train_v1.head()
```

```python
from sklearn.metrics import log_loss
import time

def pycaret_model(train, target, test, fold):
    time_start = time.time()
    
    print('setup',end=' ')
    clf_setup = setup(data=train,
        target=target,
        normalize = True,
        normalize_method = 'robust',
        silent= True,
        use_gpu=True,
        session_id = 42,
                      verbose=False,
        numeric_imputation = 'mean')
    add_metric('logloss', 'LogLoss', log_loss, greater_is_better=False, target='pred_proba')
    print(time.time()-time_start)
    
    print('create_models',end=' ')
    lst_model = []
    lst_model.append(create_model(estimator = 'lightgbm',verbose=False))
    lst_model.append(create_model(estimator = 'xgboost',verbose=False))
    lst_model.append(create_model(estimator = 'catboost',verbose=False))
    print(time.time()-time_start)
    
    print('blend_models',end=' ')
    model_blended = blend_models(estimator_list= lst_model,verbose=False)
#     pred_holdout = predict_model(model_blended)
    print(time.time()-time_start)
    
    print('finalize_model',end=' ')
    model_finalized = finalize_model(model_blended)
    print(time.time()-time_start)
    
    print('predict_proba',end=' ')
    prep_pipe = get_config('prep_pipe')
    prep_pipe.steps.append(['trained_model', model_finalized])
    pred = prep_pipe.predict_proba(test)
    print(time.time()-time_start)
    
    return pred
```

```python
result = pycaret_model(df_train_v1, target, df_test_v1, 5)
result
```
